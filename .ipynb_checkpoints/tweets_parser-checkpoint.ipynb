```python
import matplotlib.pyplot as plt
import pandas as pd

import re


def map_and_replace(mapping_dict, string):
    for element in mapping_dict:
        string = string.replace(element, mapping_dict[element])
    return string


def convert_short_day_to_full_polish(string):
    day_mapping = {
        'Mon': 'Poniedzia≈Çek',
        'Tue': 'Wtorek',
        'Wed': '≈öroda',
        'Thu': 'Czwartek',
        'Fri': 'PiƒÖtek',
        'Sat': 'Sobota',
        'Sun': 'Niedziela'
    }
    return map_and_replace(day_mapping, string)


def convert_month_name_to_num(string):
    month_mapping = {
        'Jan': '01',
        'Feb': '02',
        'Mar': '03',
        'Apr': '04',
        'May': '05',
        'Jun': '06',
        'Jul': '07',
        'Aug': '08',
        'Sep': '09',
        'Oct': '10',
        'Nov': '11',
        'Dec': '12'
    }
    return map_and_replace(month_mapping, string)


def remove_stop_words(input_string):
    polish_stop_words = [
        "i", "w", "z", "na", "do", "od", "przez", "po", "za", "za≈õ", "ale", "lub", "czy", 
        "te≈º", "jest", "byƒá", "nie", "co", "kt√≥ry", "ten", "ta", "to", "tak", "taki", "tylko", 
        "ju≈º", "teraz", "jeszcze", "ni≈º", "nad", "pod", "a", "≈ºe", "bardzo", "ci", "ciebie", 
        "cie", "co≈õ", "kto", "kt√≥re", "kt√≥rego", "kt√≥rej", "kt√≥rych", "kt√≥rzy", "je≈õli", "gdy", 
        "gdyby", "≈ºeby", "bo", "poniewa≈º", "dlatego", "wiƒôc", "tak≈ºe", "albo", "lub", "czy"
    ]
    words = input_string.split()
    filtered_words = [word for word in words if word.lower() not in polish_stop_words]
    return ' '.join(filtered_words)


df = pd.read_csv('dane7.csv')
df['created_at'] = df['created_at'].apply(convert_short_day_to_full_polish)
df['created_at'] = df['created_at'].apply(convert_month_name_to_num)
df['text_without_stopwords'] = df['text'].apply(remove_stop_words)
links_to_tweets = df['tweet_url'].to_list()
links_in_tweets = df['urls'].to_list()
links_to_media = df['media'].to_list()
```

```python
def print_top5_rows(data_frame, column_name):
    sorted_df = data_frame.sort_values(by=column_name, ascending=False)
    sorted_df.reset_index()
    print('='*100)
    print(f"Top 5 rows with highest values in '{column_name}':")
    for i in range(4):
        selected_row = sorted_df.iloc[i]
        row_dict = selected_row.to_dict()
        print('-'*100)
        print(f"Row {i}:\n{row_dict}")
    del sorted_df
    

print_top5_rows(df, 'favorite_count')
print_top5_rows(df, 'retweet_count')
```

```python
def print_bool_columns(data_frame, column_name, bool_to_print, column_to_print=False):
    data_frame[column_name] = data_frame[column_name].replace({'false': False, 'true': True})
    filtered_rows = data_frame[data_frame[column_name] == bool_to_print]
    print('='*100)
    if column_to_print is not False:
        print("Verified users id's:")
    for index, row in filtered_rows.iterrows():
        row_dict = row.to_dict()
        print('-'*100)
        if column_to_print is not False:
            if column_to_print in row_dict:
                print(f"{row_dict[column_to_print]}")
        else:
            print(f"Row {index}:\n{row_dict}")


print_bool_columns(df, 'possibly_sensitive', False)
```

```python
print_bool_columns(df, 'user_verified', True, 'user_id')
```

```python
def find_most_fresh_user_id(data_frame, time_column, user_id_column):
    data_frame[time_column] = pd.to_datetime(data_frame[time_column], format='%a %b %d %H:%M:%S +0000 %Y')
    most_fresh_user = data_frame.loc[data_frame[time_column].idxmax()]
    return most_fresh_user[user_id_column]


def find_user_id_with_most_followers(data_frame, user_id_column, followers_count_column):
    most_followers_index = data_frame[followers_count_column].idxmax()
    most_followers_user_id = data_frame.loc[most_followers_index, user_id_column]
    return most_followers_user_id


def print_user_rows(data_frame, user_id):
    user_rows = data_frame[data_frame['user_id'] == user_id]
    user_rows_list = user_rows.to_dict(orient='records')
    print('='*100)
    for i, row_dict in enumerate(user_rows_list, 1):
        print('-'*100)
        print(f"Row {i} with user_id {user_id}:\n{row_dict}")


most_fresh_user_id = find_most_fresh_user_id(df, 'user_created_at', 'user_id')
user_id_with_most_followers = find_user_id_with_most_followers(df, 'user_id', 'user_followers_count')
print_user_rows(df, most_fresh_user_id)
print_user_rows(df, user_id_with_most_followers)
```

```python
def tweets_by_day_analizer(data_frame, time_column, return_max_day=False):
    cloned_df = data_frame.copy()
    tweets_per_day_df = cloned_df[time_column].str.split(' ').str[0].value_counts().reset_index()
    tweets_per_day_df.columns = ['Day', 'Tweets']
    if return_max_day:
        max_day = tweets_per_day_df.loc[tweets_per_day_df['Tweets'].idxmax(), 'Day']
        return max_day
    else:
        return tweets_per_day_df


most_tweets_day = tweets_by_day_analizer(df, 'created_at', return_max_day=True)
print(f"Day with the most tweets: {most_tweets_day}")
```

```python
tweets_per_day_df = tweets_by_day_analizer(df, 'created_at')
print(f"Tweets per day: {tweets_per_day_df}")
plt.bar(tweets_per_day_df['Day'], tweets_per_day_df['Tweets'])
plt.xlabel('Day of the Week')
plt.ylabel('Number of Tweets')
plt.title('Number of Tweets per Day')
plt.show()
```
